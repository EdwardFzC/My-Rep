## 2 - 1% min iteraciones random

```python
# Genera iteraciones random de combinaciones de portafolios
iteraciones = 50000
max_weight = 0.94
min_weight = 0.01
portfolios = []
for _ in range(iteraciones):
    weights = np.random.dirichlet(np.ones(len(activos)), size=1)[0]
    weights = np.clip(weights, min_weight, max_weight)
    portfolios.append(weights)

# Convierte las combinaciones a un DataFrame
portfolios = pd.DataFrame(portfolios, columns=activos)

portfolios
```

```python
# Calcular el rendimiento esperado y el riesgo del portafolio
rendimiento_esperado = []
riesgo_portafolio = []
for _, row in portfolios.iterrows():
    # Calcular el rendimiento esperado del portafolio
    rendimiento_port = np.dot(row, media_rends)
    rendimiento_esperado.append(rendimiento_port)

    # Calcular el riesgo del portafolio (desviación estándar)
    varianza_port = np.dot(np.dot(row, cov_rends), row.T)
    riesgo_portafolio.append(np.sqrt(varianza_port))

# Crear el DataFrame 'portafolio'
portafolio = portfolios.copy()
portafolio['Rendimiento Esperado'] = rendimiento_esperado
portafolio['Riesgo del Portafolio'] = riesgo_portafolio

print("DataFrame 'portafolio':")
print(portafolio.head())
```

##Portafolio de minima Varianza

```python
import cvxopt as opt
from cvxopt import blas, solvers
```

```python

# Conversión a matrices cvxopt
n = len(media_rends)
S = opt.matrix(cov_rends.values)
pbar = opt.matrix(media_rends.values)

# Matrices de restricciones
G = -opt.matrix(np.eye(n))  # Matriz de identidad negativa
h = opt.matrix(0.0, (n ,1))
A = opt.matrix(1.0, (1, n))
b = opt.matrix(1.0)

# Resolver el problema de optimización cuadrática
solvers.options['show_progress'] = False
sol = solvers.qp(S, opt.matrix(0.0, (n, 1)), G, h, A, b)

# Proporciones del portafolio de mínima varianza
proporciones = np.array(sol['x']).flatten()

# Cálculo del rendimiento esperado, riesgo y ratio de Sharpe
rendimiento_esperado_min_var = np.dot(proporciones, media_rends)
riesgo_min_var = np.sqrt(np.dot(proporciones.T, np.dot(cov_rends, proporciones)))
tasa_libre_riesgo = .11/12
ratio_sharpe_min_var = (rendimiento_esperado_min_var - tasa_libre_riesgo) / riesgo_min_var

```

# Linea CML

```python
# Encontrar el portafolio tangente
portafolio_tangente = portafolio.loc[portafolio['Ratio de Sharpe'].idxmax()]

# Calcular la CML
def cml(riesgo, rf, rm, sr):
    return rf + sr * riesgo

cml_riesgos = np.linspace(0, max(portafolio['Riesgo del Portafolio']), 100)
cml_rendimientos = cml(cml_riesgos, tasa_libre_riesgo, portafolio_tangente['Rendimiento Esperado'], portafolio_tangente['Ratio de Sharpe'])

```

```python
# Calcular el valor de la inversión a lo largo del tiempo para cada activo y portafolio
inversion_inicial = 100
proporciones_tangente = portfolios.loc[portafolio['Ratio de Sharpe'].idxmax()]
proporciones_min_var = np.array(sol['x']).flatten()

def calcular_valor_inversion(rend_acum, inversion_inicial):
    return inversion_inicial * (1 + rend_acum)

valores_activos = {activo: calcular_valor_inversion(rends_acum[activo], inversion_inicial) for activo in activos}

# Calcular el valor de la inversión para los portafolios tangente y de mínima varianza
valores_portafolio_tangente = inversion_inicial * (1 + np.dot(rends_acum, proporciones_tangente))
valores_portafolio_min_var = inversion_inicial * (1 + np.dot(rends_acum, proporciones_min_var))

```

```python
def calcular_drawdown(valores):
    max_valor_acumulado = np.maximum.accumulate(valores)
    drawdown = (valores - max_valor_acumulado) / max_valor_acumulado * 100
    return drawdown

drawdown_tangente = calcular_drawdown(valores_portafolio_tangente)
drawdown_min_var = calcular_drawdown(valores_portafolio_min_var)
```

```python
# Definir las proporciones del portafolio de pesos iguales
num_activos = len(activos)
proporciones_igual = np.ones(num_activos) / num_activos

# Calcular el rendimiento esperado, riesgo y ratio de Sharpe del portafolio de pesos iguales
rendimiento_esperado_igual = np.dot(proporciones_igual, media_rends)
riesgo_igual = np.sqrt(np.dot(proporciones_igual.T, np.dot(cov_rends, proporciones_igual)))
ratio_sharpe_igual = (rendimiento_esperado_igual - tasa_libre_riesgo) / riesgo_igual

# Calcular el valor de la inversión para el portafolio de pesos iguales
valores_portafolio_igual = inversion_inicial * (1 + np.dot(rends_acum, proporciones_igual))

# Calcular el drawdown del portafolio de pesos iguales
drawdown_igual = calcular_drawdown(valores_portafolio_igual)

riesgo_libre=0
```

```python
# Definir el rendimiento esperado del mercado (S&P 500)
rendimiento_esperado_mercado = media_rends['Syp500_r']

# Calcular Beta para cada portafolio en relación con S&P 500
def calcular_beta(proporciones, cov_rends, media_rends, rendimiento_esperado_mercado):
    # Covarianza entre el portafolio y el mercado
    cov_portafolio_mercado = np.dot(proporciones, cov_rends['Syp500_r'])
    # Varianza del mercado
    var_mercado = cov_rends.loc['Syp500_r', 'Syp500_r']
    # Beta del portafolio
    beta = cov_portafolio_mercado / var_mercado
    return beta

beta_tangente = calcular_beta(proporciones_tangente, cov_rends, media_rends, rendimiento_esperado_mercado)
beta_min_var = calcular_beta(proporciones_min_var, cov_rends, media_rends, rendimiento_esperado_mercado)
beta_igual = calcular_beta(proporciones_igual, cov_rends, media_rends, rendimiento_esperado_mercado)

# Calcular el CAPM para cada portafolio
def calcular_capm(beta, tasa_libre_riesgo, rendimiento_esperado_mercado):
    return tasa_libre_riesgo + beta * (rendimiento_esperado_mercado - tasa_libre_riesgo)

capm_tangente = calcular_capm(beta_tangente, tasa_libre_riesgo, rendimiento_esperado_mercado)
capm_min_var = calcular_capm(beta_min_var, tasa_libre_riesgo, rendimiento_esperado_mercado)
capm_igual = calcular_capm(beta_igual, tasa_libre_riesgo, rendimiento_esperado_mercado)
```
