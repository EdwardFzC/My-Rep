```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Ejemplo de datos de precios de Bitcoin
data = {
    'Date': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'],
    'Price': [29000, 33000, 45000, 59000, 50000]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], label='Bitcoin Price', color='blue')

# Configurar la escala logarítmica en el eje Y
plt.yscale('   ')

# Añadir etiquetas y título
plt.xlabel('Date')
plt.ylabel('Price (log scale)')
plt.title('Bitcoin Price Evolution')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
# Crear la gráfica
fig = px.line(df, x='Date', y='Price', title='Bitcoin Price Evolution')

# Configurar la escala logarítmica en el eje Y
fig.update_layout(yaxis_type="log")

# Mostrar la gráfica
fig.show()
```
